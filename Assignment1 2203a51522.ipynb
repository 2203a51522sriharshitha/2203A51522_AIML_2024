Part-02: Lab task is to visualize the datasets or data using Matplot function graphs

```python

import matplotlib.pyplot as plt

people = ['sri', 'prasanna', 'snithika', 'sathvika', 'sathvik', 'varshitha', 'varshini', 'shaistha', 'charan', 'tharun', 'rajesh']
height = [20,21,13,14,15,16,17,18,19,10,11]
weight = [34,54,23,87,12,56,90,25,65,34,87]
plt.scatter(weight, height)
plt.title("relationship between weight and height of patients")
plt.ylabel("weight of patients in kgs")
plt.show()
```

part-2 Generate Bar plot

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,14))
plt.title("people's weight in kgs", fontsize=16, fontstyle='italic')

plt.bar(x=people, height=weight, width=0.6, color='orange', edgecolor='k',alpha=0.6)

plt.xlabel("people",fontsize=15)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight9in(kgs)",fontsize=15)
plt.show()
```

Generate Histogram

```python
import numpy as np
plt.figure(figsize=(7,5))

plt.hist(weight, color='red', edgecolor='k', alpha=0.75, bins=5)

plt.title("Histogram of patient weight", fontsize=15)
plt.xlabel("weight in kgs",fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
